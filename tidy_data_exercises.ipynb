```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import new_lib as nl
```

```python
url = nl.get_db_url('tidy_data')
```

```python
query = 'SELECT * FROM attendance'
attendance = pd.read_sql(query, url)
attendance
```

```python
attendance = attendance.rename(columns = {'Unnamed: 0': 'Name'})
attendance.T
```

```python
attendance = attendance.melt(id_vars = 'Name', var_name = 'Date', value_name = 'Status')
attendance
```

```python
attendance['Status'] = attendance['Status'].map({'P': 1, 'T': .9, 'A': 0, 'H': .5})
```

```python
attendance.groupby('Name').mean()
```
